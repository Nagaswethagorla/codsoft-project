```python
import pandas as pd
import numpy as np
import csv
dataframe1 = pd.read_csv("titanic.csv", header=None)
temprows = []
for x in dataframe1[0]:
    temprows.append(next(csv.reader([x])))

new_df = pd.DataFrame(temprows)
new_df.columns = new_df.iloc[0]
new_df= new_df[1:].reset_index(drop=True)

dataFrame = new_df[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]#considering only required columns
dataFrame = dataFrame.replace('', np.nan)

dataFrame['Age'] = dataFrame['Age'].fillna(0)
dataFrame['Fare'] = dataFrame['Fare'].fillna(0)
dataFrame['Embarked'] = dataFrame['Embarked'].fillna(dataFrame['Embarked'].mode()[0])

num_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
for col in num_cols:
    dataFrame[col] = pd.to_numeric(dataFrame[col], errors='coerce')

dataFrame['Age'] = dataFrame['Age'].fillna(0)
dataFrame['Fare'] = dataFrame['Fare'].fillna(0)
dataFrame['Sex'] = dataFrame['Sex'].map({'male': 1, 'female': 0})
dataFrame['Embarked']=dataFrame['Embarked'].map({"S": 0, "C": 1, "Q": 2})

x_data = dataFrame.drop('Survived', axis=1)#input data
y_data = dataFrame['Survived']#target data

from sklearn.model_selection import train_test_split
x_train_data, x_test_data, y_train_data, y_test_data = train_test_split(x_data, y_data, test_size=0.2, random_state=42)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)

model.fit(x_train_data, y_train_data)
prediction = model.predict(x_test_data)

from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test_data, prediction))
```
